```python
import numpy as np
from math import factorial
import matplotlib.pyplot as plt
```

```python
bigO = [
    (lambda x: np.log(x),'log(n)'),
    (lambda x: np.sqrt(x),'sqrt(n)'),
    (lambda x: x,'n'),
    (lambda x: x*np.log(x),'n log(n)'),
    (lambda x: np.power(x,2),'n^2'),
    (lambda x: np.power(x,3),'n^3'),
    (lambda x: np.power(2,x),'2^n'),
    (lambda x: list(map(factorial,x)),'n!')
]


plt.figure(figsize=(8,5))

n = range(2,13)
for complexity in bigO:
    plt.semilogy(n,complexity[0](n),'o-',linewidth=3)
    
plt.legend([x[1] for x in bigO])
plt.savefig('img/algorithms_time_complexity.png')
```
