## Ramsey Numbers

The Ramsey theorem states that you will find monochromatic cliques in any edge labeling for a sufficiently large graph. 

For the most common example of two different colors (say, blue and red), the integer $v = R(n,m)$ is the size of the graph where any possible labeling will result in either a clique of $n$ red or $m$ blue, or vice versa. 

In general, the Ramsey numbers are known for only very few values of $n$ and $m$. The special case of $n=3$ and $m=3$ is popular as an interview question: 

"*At a party of 6 people, show that either at least 3 people know each other or at least 3 people don't know each other.*"

#### R(2,m) = m

Consider a Graph of size m. All nodes are connected by edges that are either red or blue. If any two nodes are connected by a red edge, then there is a clique of $n$=2 red nodes. If no two nodes are connected by a red edge then, by exclusion, all of the $m$ nodes connected by blue edges, and there is a clique of $m$ blue nodes. 

#### R(3,3) = 6

Consider a Graph of size 6. Each node is connected to 5 others via edges. Since there are only two colors, each node has at minimum 3 edges of the same color (say: blue). 


We can think of those edges as defining a subgraph of size at least $m$ with $3\leq m \leq 5$.

If *at least one* of the $m$ nodes in that subgraph are connected with a blue edge, then at *least 3 nodes* are connected with blue edges. 

If *none* of the $m$ nodes in that subgraph are connected with a blue edge, then $m\geq 3$ nodes are connected with red edges.


#### R(4,3) = 9

Consider a Graph of size 9. Each node is connected to 8 others via edges. Since there are only two colors, each node has at minimum 4 edges of the same color (say: blue). 


We can think of those edges as defining a subgraph of size at least $m$ with $4\leq m \leq 8$.

If *any* of the $m$ nodes in that subgraph are connected with a blue edge, then three nodes are connected with blue edges. 

If *none* of the $m$ nodes in that subgraph are connected with a blue edge, then $m\geq4$ nodes are connected with red edges.


```python
import networkx as nx
import matplotlib.pyplot as plt


G=nx.Graph()


def get_edges(n):
    """
    Yield all edges of complete graph as tuples
    """
    for i in range(1,n):
        for j in range(i+1,n+1):
            yield (i,j)

            
edges = get_edges(n=6)

G.add_edges_from(edges)
pos = nx.spring_layout(G)
G.remove_edges_from(G.edges)

G.add_edges_from([(1,2),(1,3),(1,4)])
        
plt.figure(figsize=(15,5))

plt.subplot(131)
nx.draw(G,pos,node_color='#A0CBE2',edge_color='blue',width=4,edge_cmap=plt.cm.Blues,with_labels=True)


plt.subplot(132)

G.add_edges_from([(2,4),(2,3),(3,4)])
nx.draw(G,pos,node_color='#A0CBE2',edge_color=3*['blue']+3*['red'],width=4,edge_cmap=plt.cm.Blues,with_labels=True)

plt.subplot(133)
nx.draw(G,pos,node_color='#A0CBE2',edge_color=4*['blue']+2*['red'],width=4,edge_cmap=plt.cm.Blues,with_labels=True)
```

```python
import networkx as nx

G=nx.Graph()


def get_edges(n):
    """
    Yield all edges of complete graph as tuples
    """
    for i in range(1,n):
        for j in range(i+1,n+1):
            yield (i,j)

            
edges = get_edges(n=9)

G.add_edges_from(edges)
pos = nx.spring_layout(G)
G.remove_edges_from(G.edges)

G.add_edges_from([(1,i) for i in range(2,6)])
        
plt.figure(figsize=(15,5))

plt.subplot(131)
nx.draw(G,pos,node_color='#A0CBE2',edge_color='blue',width=4,edge_cmap=plt.cm.Blues,with_labels=True)


plt.subplot(132)

for i in range(2,5):
    G.add_edges_from([(i,j) for j in range(i+1,6)])
nx.draw(G,pos,node_color='#A0CBE2',edge_color=4*['blue']+6*['red'],width=4,edge_cmap=plt.cm.Blues,with_labels=True)

plt.subplot(133)
nx.draw(G,pos,node_color='#A0CBE2',edge_color=5*['blue']+5*['red'],width=4,edge_cmap=plt.cm.Blues,with_labels=True)
```
