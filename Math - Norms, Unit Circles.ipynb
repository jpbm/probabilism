```python
import matplotlib.pyplot as plt
import numpy as np


plt.rcParams.update({
    'axes.spines.bottom': False,
    'axes.spines.left': False,
    'axes.spines.right': False,
    'axes.spines.top': False,
    'ytick.left': False,
    'xtick.bottom': False,
    'ytick.labelleft': False,
    'xtick.labelbottom': False
})


def unit_circle(q):
    
    N=200
    x=np.linspace(0,1,N)
    
    sec1 = (1-np.abs(x)**q)**(1/q)
    
    y = np.hstack([sec1,-sec1[::-1],-sec1,sec1[::-1]])
    x = np.hstack([x,x[::-1],-x,-x[::-1]])
    
    return x,y


fig,ax = plt.subplots(3,3,figsize=(8,8))

Q = [0.0001,0.5,1,1.5,2,2.5,3,3.5,100]
for q,a in zip(Q,ax.flatten()):
    x,y = unit_circle(q)
    a.plot(x,y)
    #a.text(x=0.6,y=0.75,s=str(q),fontsize=12)
    a.set_xlabel('$L^{%1.1f}$' % q,fontsize=17)
    
    if q < 0.1:
        a.set_xlabel('$L^{%1.4f}$' % q,fontsize=17)
        
    if q%1 ==0:
        a.set_xlabel('$L^{%i}$' % q,fontsize=17)
        
    
    
plt.savefig('img/unitsets.png')
```
